```python
#importacion de libraries
import pandas as pd 
```

```python
### IMPORTO EL DATAFRAME DE TRAIN PARA QUITARLE LOS FEATURES NULOS
df_train=pd.read_csv('join_features_categoricos_train_final.csv')
df_train.shape
```

```python
features_to_remove = (df_train != 0).any(axis=0)
```

```python
#ELIMINO FEATURES CON NULOS EN EL ARCHIVO DE TRAIN
df_train = df_train.loc[:,features_to_remove]
print df_train.shape
df_train.to_csv('df_no_null_features_train.csv')
df_train.head()
```
